```python
for col in df.columns:
    if df[col].dtype!="object":
        plt.boxplot(df[col])
        plt.xlabel(col)
        plt.show()
```

```python
# out_list=[]
# for col in df.columns:
#     if df[col].dtype!="object":
#         out_list.append(col)
```

```python
out_list=['age','fnlwgt','education-num','hours-per-week']
```

```python
out_list
```

```python
for col in out_list:
    
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    UW=Q3+1.5*IQR
    LW=Q1-1.5*IQR
    df=df[(df[col]>=LW) & (df[col]<=UW)]
```

```python
for col in out_list:
    plt.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```
